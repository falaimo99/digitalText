```python
from bs4 import BeautifulSoup

path = './compagne.xml'

with open(path, 'r', encoding='utf-8') as f:

    soup = BeautifulSoup(f, features='xml')
    persName = soup.find_all('persName')
    place = soup.find_all('place')

for i in place:
    print(i['id'])

for place in place:
    id = place['id']
    newPersNameId = (str(id))[:1].upper() + (str(id))[1:]
    place['id'] = newPersNameId

for person in persName:
    id = person['id']
    newPersNameId = (str(id))[:2].upper() + (str(id))[2:]
    person['id'] = newPersNameId

with open('compagne.xml', 'w', encoding='utf-8') as f:
    f.write(str(soup))
    
```
