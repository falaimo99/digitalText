```python
from bs4 import BeautifulSoup
from SPARQLWrapper import SPARQLWrapper, JSON
import sparql_dataframe
```

```python
with open('.\produced_data\header.xml', 'r', encoding='utf-8') as f:
    soup = BeautifulSoup(f, features='xml')
    works_of_art = {}
    # works = soup.find_all('item')
    # sameAs = [work['sameAs'] for work in works]
    for i in soup.find_all('item'):
        if 'sameAs' in i.attrs.keys():
            sameAs = i['sameAs']
            id = i['xml:id']
            wd_entity = sameAs.split('/entity/')
            works_of_art[wd_entity[1]] = id

print(works_of_art)
len(works_of_art) 
```

```python
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")

for i in works_of_art:

    endpoint = 'https://query.wikidata.org/sparql'
    
    querywork = """
   #defaultView:ImageGrid{"hide": ["?image"]}
SELECT ?thing ?thingLabel ?image (URI(CONCAT('https://tools.wmflabs.org/mirador/?manifest=', STR(?iiif))) AS ?iiiflink) WHERE {
wd:%s wdt:P6108 ?iiif;
   wdt:P170 ?author. # creator: Turner
?thing wdt:P6108 ?iiif.
OPTIONAL{ ?thing wdt:P18 ?image }
SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
    """%(i)
    
    df = sparql_dataframe.get(endpoint, querywork, post='True')
    print(df)
```
